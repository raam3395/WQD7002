```python
import numpy as np
import pandas as pd
import os

#For data visualization
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
%matplotlib inline
import plotly.express as px
import seaborn as sns

```

```python
unifi = pd.read_csv("unifidone.csv")
maxis = pd.read_csv("maxisdone.csv")
celcom = pd.read_csv("celcomdone.csv")
digi = pd.read_csv("digidone.csv")
umobile = pd.read_csv("umobiledone.csv")
```

```python

def change_label(df):
    df['Sentiment'] = df['Sentiment'].map({-1:'Negative', 0:'Neutral', 1:'Positive'})

change_label(digi)
change_label(maxis)
change_label(unifi)
change_label(celcom)
change_label(umobile)
```

```python
from wordcloud import WordCloud,ImageColorGenerator
from PIL import Image
import urllib
import requests
import matplotlib.pyplot as plt 

def wordCloud(tweet,x) :
    
    # combining the image with the dataset
    Mask = np.array(Image.open(requests.get('http://clipart-library.com/image_gallery2/Twitter-PNG-Image.png', stream=True).raw))

    # We use the ImageColorGenerator library from Wordcloud 
    # Here we take the color of the image and impose it over our wordcloud
    image_colors = ImageColorGenerator(Mask)

    # Now we use the WordCloud function from the wordcloud library 
    wc = WordCloud(background_color='black', height=1600, width=4000,mask=Mask).generate(" ".join(tweet[tweet.Sentiment == x].processed_tweets))
    #join(data[data.polarity == 0].processed_tweets)

    # Size of the image generated 
    plt.figure(figsize=(10,20))

    # Here we recolor the words from the dataset to the image's color
    # recolor just recolors the default colors to the image's blue color
    # interpolation is used to smooth the image generated 
    plt.imshow(wc.recolor(color_func=image_colors),interpolation="hamming")

    plt.axis('off')
    return plt.show()
```

```python
# TM
fig = px.pie(unifi, names='Sentiment', title ='Pie chart of Telekom Malaysia sentiments of tweets')
fig.show()
```

```python
# The distribution of sentiments TM
plt.figure(figsize=(12,6))
sns.countplot(x='Sentiment',data=unifi).set_title('Telekom Malaysia Sentiment Frequency Plot')
```

```python
wordCloud(celcom,1) 
```

```python
def scoretelco(telco) :
    score = telco['Sentiment'].mean() * 100
    return round(score,2)
```

```python
print("TM : ", scoretelco(unifi), "%")
print("Maxis : ", scoretelco(maxis), "%")
print("Celcom : ", scoretelco(celcom), "%")
print("Digi : ", scoretelco(digi), "%")
print("Umobile : ", scoretelco(umobile), "%")
```
